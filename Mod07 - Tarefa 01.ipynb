````python
categorias = pd.DataFrame({'variavel' : tipos.index,
                          'tipo' : tipos.values})
categorias
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

Transformando a variável sexo em binária utilizando map. 1 indica masculino e 0 feminino

````python
df.sexo.unique()
````

````python
df.sexo = df.sexo.map({'M' : 1, 'F' : 0})
df
````

Transformando as variaveis posse_de_veiculo e posse_do_imovel em binarios

````python
df.posse_de_veiculo = df.posse_de_veiculo.map({'Y' : 1, 'N' : 0})
````

````python
df.posse_de_imovel = df.posse_de_imovel.map({'Y' : 1, 'N' : 0})
````

````python
df_encoded = df.drop(['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'],axis = 1)
df_encoded.head()
````

````python
df_encoded.mau = df_encoded.mau.astype(int)
````

````python
df_encoded.dtypes
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
tab = pd.crosstab(df_encoded['possui_email'], df_encoded['posse_de_veiculo'])
tab
````

Com certeza a variavel 'posse_de_veiculo' tem mais influencia sobre a 'mau', pois está relacionada o poder aquisitivo da pessoa.
